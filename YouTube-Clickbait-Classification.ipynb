```python
# Intake training and video data
# This code was run one time with all video data saved to the working directory
def download(youtube_url):
    ydl_opts = {
        'format': '160',
        'outtmpl': 'videos/%(id)s.%(ext)s'
    }
    
    with youtube_dl.YoutubeDL(ydl_opts) as ydl:
        ydl.download([youtube_url])

df = pd.read_csv("train.csv")

if not os.path.exists("videos"):
  os.mkdir("videos")

error_url_list = []

for i in range(df.shape[0]):
  if i%300 == 0:
    time.sleep(30)

  try:
    download(df.loc[i,"URL"])
  
  except Exception as e:
    print("Exception: ", e)
    error_url_list.append(df.loc[i,"URL"])
    continue

if len(error_url_list) > 0:
  print(">>>List below shows the URL of missing videos:")
  for url in error_url_list:
    print(url)
```

```python
# Format training and testing data
train = pd.read_csv("train.csv", usecols = ["class", "viewCount", "likeCount","dislikeCount","commentCount","title", "description", "user_comment_1", "user_comment_2", "user_comment_3", "user_comment_4", "user_comment_5", "user_comment_6", "user_comment_7", "user_comment_8", "user_comment_9", "user_comment_10"])
test = pd.read_csv("test_1.csv", usecols = ["ID", "viewCount", "likeCount","dislikeCount","commentCount","title", "description", "user_comment_1", "user_comment_2", "user_comment_3", "user_comment_4", "user_comment_5", "user_comment_6", "user_comment_7", "user_comment_8", "user_comment_9", "user_comment_10"])
print(train.head())
```

```python
# Generate variable proportion of likes to views and dislikes to views
train['like_prop'] = train['likeCount']/train['viewCount']
train['dislike_prop'] = train['dislikeCount']/train['viewCount']

test['like_prop'] = test['likeCount']/test['viewCount']
test['dislike_prop'] = test['dislikeCount']/test['viewCount']
```

```python
# Generate proportion of comments to views
train['comment_prop'] = train['commentCount']/train['viewCount']

test['comment_prop'] = test['commentCount']/test['viewCount']
```

```python
# Compare more likes or more dislikes on the video
train['difference'] = train['likeCount'] - train['dislikeCount']
train['more_likes'] = train['difference'].apply(lambda x: 1 if x>0 else 0)
train= train.drop('difference', 1)
print(sum(train.more_likes))

test['difference'] = test['likeCount'] - test['dislikeCount']
test['more_likes'] = test['difference'].apply(lambda x: 1 if x>0 else 0)
test= test.drop('difference', 1)
print(sum(test.more_likes))
```

```python
# Prep data for modeling and cast all text values to numeric as shown in baseline code sample

y_train = train["class"]

x_train = train.drop("class", axis=1)
x_test = test.drop("ID", axis=1)

x_train["title"] = x_train["title"].apply(lambda x: len(x))
x_test["title"] = x_test["title"].apply(lambda x: len(x))

x_train["description"] = x_train["description"].apply(lambda x: len(x))
x_test["description"] = x_test["description"].apply(lambda x: len(x))

x_train["user_comment_1"] = x_train["user_comment_1"].apply(lambda x: len(x))
x_test["user_comment_1"] = x_test["user_comment_1"].apply(lambda x: len(x))

x_train["user_comment_2"] = x_train["user_comment_2"].apply(lambda x: len(x))
x_test["user_comment_2"] = x_test["user_comment_2"].apply(lambda x: len(x))

x_train["user_comment_3"] = x_train["user_comment_3"].apply(lambda x: len(x))
x_test["user_comment_3"] = x_test["user_comment_3"].apply(lambda x: len(x))

x_train["user_comment_4"] = x_train["user_comment_4"].apply(lambda x: len(x))
x_test["user_comment_4"] = x_test["user_comment_4"].apply(lambda x: len(x))

x_train["user_comment_5"] = x_train["user_comment_5"].apply(lambda x: len(x))
x_test["user_comment_5"] = x_test["user_comment_5"].apply(lambda x: len(x))

x_train["user_comment_6"] = x_train["user_comment_6"].apply(lambda x: len(x))
x_test["user_comment_6"] = x_test["user_comment_6"].apply(lambda x: len(x))

x_train["user_comment_7"] = x_train["user_comment_7"].apply(lambda x: len(x))
x_test["user_comment_7"] = x_test["user_comment_7"].apply(lambda x: len(x))

x_train["user_comment_8"] = x_train["user_comment_8"].apply(lambda x: len(x))
x_test["user_comment_8"] = x_test["user_comment_8"].apply(lambda x: len(x))

x_train["user_comment_9"] = x_train["user_comment_9"].apply(lambda x: len(x))
x_test["user_comment_9"] = x_test["user_comment_9"].apply(lambda x: len(x))

x_train["user_comment_10"] = x_train["user_comment_10"].apply(lambda x: len(x))
x_test["user_comment_10"] = x_test["user_comment_10"].apply(lambda x: len(x))
```

```python
# Normalization
x_train = x_train.apply(lambda x: (x - np.min(x)) / (np.max(x) - np.min(x)))
x_test = x_test.apply(lambda x: (x - np.min(x)) / (np.max(x) - np.min(x)))
```

```python
# Create margin hyperparameters to be explored
param_grid = {'C': [1, 10, 100, 1000], 'kernel': ['linear', 'rbf'], 'gamma':[0.01, 0.1, 0.3, 1, 5, 10]}
```

```python
# Instantiate stratified K-fold
cv = StratifiedKFold(n_splits=5, random_state=2, shuffle=True)
```

```python
# Apply grid search to find the best SVC model
grid_search = GridSearchCV(estimator = SVC(), param_grid=param_grid, cv = cv, 
                           scoring = 'accuracy', verbose = 1,
                           n_jobs = -1, return_train_score = True)
grid_search.fit(x_train, y_train)
```

```python
# View results in order of accuracy performance
results = pd.DataFrame(grid_search.cv_results_).sort_values('mean_test_score', ascending = False)
results.head()
```

```python
# Save the best SVC model
bestmodel = grid_search.best_estimator_
```

```python
# Fit the best SVC model
bestmodel.fit(x_train, y_train)
```

```python
# Make predictions with the best SVC model
train_pred = bestmodel.predict(x_train)
y_pred = bestmodel.predict(x_test)
```

```python
# View training F1 score
f1_score(y_train, train_pred)
```

```python
# Try a decision tree classifier 
parameters = {'max_depth':range(3,20), 'min_samples_split':[2, 3, 4], 'min_samples_leaf':[1, 2, 3], 'max_features': ['auto', 'sqrt', 'log2']}
```

```python
# Apply grid search to find the best decision tree classifier model
grid_search = GridSearchCV(estimator = DecisionTreeClassifier(), param_grid=parameters, cv = cv, 
                           scoring = 'accuracy', verbose = 1,
                           n_jobs = -1, return_train_score = True)
grid_search.fit(x_train, y_train)
```

```python
# View results in order of accuracy performance
results = pd.DataFrame(grid_search.cv_results_).sort_values('mean_test_score', ascending = False)
results.head()
```

```python
# Save and fit tree model
besttree = grid_search.best_estimator_
besttree.fit(x_train, y_train)
```

```python
# Make tree predictions
y_pred = besttree.predict(x_test)
```

```python
# This code chunk formats the predictions to an appropriate submission file for Kaggle
test["class"] = y_pred
test["class"] = test["class"].map(lambda x: "True" if x==1 else "False")
result = test[["ID","class"]]
result.to_csv("predictions.csv", index=False)
result.head()
```
